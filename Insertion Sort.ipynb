# Insertion Sort

Time Complexity: O(n) = n^2

Insertion sort works by starting at the left side of the list. The first element is marked as sorted by default. The second element is compared. If they need to be swapped, they are. Now the first two elements are sorted. The third element is compared with the 2nd, and if it is swapped, it also compared with the 1st. Then the first 3 elements. This process repeats for the length n of the array. A worst case scenario would be a list of complete descending order, and this would result in every element having to be compared with every other element. This results in n^2





# Python Implementation:

```python
array = [100,99,98,97,96]

def insertion_sorter(array):
    '''Returns Array Sorted via Insertion Sort'''
    for index in range (1,len(array)):#We are starting at the second element.
        value = array[index] # the value of the current sort
        i = index-1 #i is the previous index
        while i >= 0: #while i is the first index and up
            if value < array[i]: #if the current value is smaller than the previous(unsorted)
                array[i+1] = array[i] # current spot consumes previous spot's value
                array[i] = value #previous spot consumes current value
                i = i -1 # this will set i down to check for the number previous to the previous number itself. 
            else:
                break #if the order is in place, it will move on to the next index.
                
                
    return array

print(insertion_sorter(array))
```
