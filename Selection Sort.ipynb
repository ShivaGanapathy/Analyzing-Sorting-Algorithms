# Selection Sort

Time Complexity: O(n) = n^2

Selection sort works by starting at the left side of the list. We iterate through the entire list and find the minimum value. The minimum value then swaps places with the element at the beginning of the array. We then repeat the process, this time only iterating through the unsorted elements. This process results in a time complexity of n^2.




# Python Implementation:

```python
array = [100,99,98,97,96]

def selection_sorter(array):
    '''Returns Array Sorted via Selection Sort'''
    for i in range(0,len(array)-1):
        min_index = i
        for j in range(i+1,len(array)):
            if array[j] < array[min_index]:
                min_index = j
        if min_index != i:
            old_min = array[i]
            array[i] = array[min_index]
            array[min_index]=old_min 
            
    return array
        
print(selection_sorter(array))
```
