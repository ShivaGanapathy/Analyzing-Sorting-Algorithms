# Bubble Sort

Time Complexity: O(n) = n^2

Bubble sort starts by comparing consecutive elements at the beginning of the list. The first two elements are compared and switched if neccessary. Then, we compare the second and third elements and switch their spots if neccessary. This process repeats until we compare the nth and n-1th elements of the array. The last element is then marked as sorted and we go back to compare the 1st element with the 2nd. Since the last element is marked as sorted, we compare up to the n-2th and n-1th element. This process results in a time complexity of n^2.




# Python Implementation:

```python
array = [100,99,98,97,96]

def bubble_sorter(array):
    '''Returns Array Sorted via Bubble Sort'''
    for i in range (0,len(array)-1): 
        for j in range (0,len(array)-1-i): # the minus i ensures that we do not sort the already sorted part.
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j] #this switches places
                
    return array

print(bubble_sorter(array))
```
